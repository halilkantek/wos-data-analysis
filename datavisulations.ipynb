```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel('h.xlsx')
```

```python
df.info()
```

```python
df.describe().style.background_gradient(cmap='twilight')
```

```python
df.isna().sum()/len(df)*100
```

```python
colors=['#fb8500','#ffc2d1','#ae2012','#00afb9','#6a994e','#e63946','#ffb703','#7fff00']
# 'Country' sütununu kullanarak çubuk grafiği çiz
df['country'].value_counts().plot(kind='bar', figsize=(15, 5), color=colors)

# Grafiği düzenleme

plt.xlabel("Countries")
plt.ylabel("Count")

# Grafiği göster
plt.show()
```
